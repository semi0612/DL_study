```python
exam = pd.read_excel("excel_exam.xlsx")
# pandas의 query()를 이용해 조건 하나 걸기
# 수학점수가 50 이상인 경우 출력
exam.query('math >= 50')
```

```python
# 조건 두개 걸기
# 수학점수가 50이상 영어점수는 90 초과
exam.query('math >= 50 and english > 90')
```

```python
# 1반이고 수학이 50점 이상
exam = exam.rename({'class':'nclass'}, axis=1)
```

```python
exam.query('nclass == 1 and math >= 50')
```

```python
# nclass를 기준으로 오름차순 정렬한 다음
# 정렬된 기준(각 반)에서 수학점수를 기준으로 오름차순 정렬
exam.sort_values(['nclass', 'math'])
```

```python
# 변수별로 정렬 순서를 다르게 지정하려면
# `ascending`도 []를 사용해 묶어주면 된다
exam.sort_values(['nclass', 'math'], ascending=[True, False])
```

```python
exam.assign(total = exam['math']+exam['english']+exam['science'])
```

```python
# 수학 평균 구하기
# 행 이름 = (계산할 컬럼, '어떤 값을 얻고 싶은가')
exam.agg(mean_1 = ('math', 'mean'))
```

```python
# 각 집단별로 묶어서 각각에 대하여 최솟값
print(exam.groupby('nclass', as_index=False).min())
#    nclass  id  math  english  science
# 0       1   1    30       86       50
# 1       2   5    25       78       25
# 2       3   9    20       65       15
# 3       4  13    46       56       12
# 4       5  17    65       68       58
```

```python
# 만약 구해진 값을 모두 보지 않고 특정 열만 보고싶다면
exam.groupby('nclass').min()[['math', 'science']]
```

```python
# 혹은 전체값은 필요없고 몇 열만 필요한 경우라면
exam[['nclass', 'math', 'science']].groupby('nclass').min()
```

```python
# \ = 이어주는 기호. 이래 위는 한줄이다라는 표시
# nclass 별로 먼저 분리하고
# math 평균을 구했다.
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'))
```

```python
# 만약 여러 통계랑을 한번에 구하고 싶다면
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'),
        sum_math = ('math', 'sum'))
```

```python
# 가로로 합치기
# 데이터 만들기
test1 = pd.DataFrame({'id' : [0, 1, 2, 3, 6],
                     'midterm' : [60, 80, 70, 90, 95]})
test2 = pd.DataFrame({'id' : [1, 2, 3, 5, 7],
                     'final' : [70, 83, 65, 95, 80]})
```

```python
total = pd.merge(test1, test2, how = 'left', on = 'id')
print(total)
#    id  midterm  final
# 0   1       60     70
# 1   2       80     83
# 2   3       70     65
# 3   4       90     95
# 4   5       95     80
```

```python
# 다른 데이터를 활용해 변수 추가하기
teacher = pd.DataFrame({'nclass' : [1, 2, 3, 4, 5],
                       'teacher' : ['kim', 'lee', 'park', 'choi', 'jung']})
exam_new = pd.merge(exam, teacher)
exam_new
```

```python
print(pd.concat([test1, test2], join = 'inner'))
#    id
# 0   0
# 1   1
# 2   2
# 3   3
# 4   6
# 0   1
# 1   2
# 2   3
# 3   5
# 4   7
```

```python
test3 = test2.rename({'final':'midterm'}, axis=1)
test3
```

```python
# test2의 final 컬럼명을 midterm 으로 바꾸고 test3라고 명명해라
test3 = test2.rename({'final':'midterm'}, axis=1)
# 만들어진 test3와 test1을 붙여라. 기준열은 test1이다
group_test = pd.concat([test1, test3])
# id 별로 group_test의 평균값을 구해라
group_test.groupby('id')['midterm'].mean()
group_test.groupby('id').agg(mean_t = ('midterm', 'mean'))
group_test.groupby('id').agg(np.mean)
```

```python
# 결측치 만들기
exam.iloc[[0, 1, 2], [2, 3, 4]] = np.nan
# 결측치 확인하기
df = exam.copy()
print(df)
pd.isna(df).sum()
# id         0
# nclass     0
# math       3
# english    3
# science    3
# dtype: int64
```

```python
# 만약 수학 열에 있는 결측치만 제거하고 싶다면
print(df.dropna(subset='math'))
# 만약 df.dropna()에 변수를 지정해 주지 않는다면
# 모든 변수에서 결측치가 없는 행만 추출한다
df.dropna()
```

```python
# 각각의 칼럼에 대해서 평균값 구해보기
df.apply(np.mean)
# id         10.500000
# nclass      3.000000
# math       58.470588
# english    83.352941
# science    58.882353
# dtype: float64
```

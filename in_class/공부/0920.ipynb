```python
from pydataset import data
mpg = data("mpg")
```

```python
mpg.info()
```

```python
# countplot로 drv 컬럼을 cyl별로 그려봅시다
!pip install seaborn
import seaborn as sns
sns.countplot(data = mpg, x = 'drv', hue='cyl')
```

```python
# total 열을 cty 와 hwy의 평균으로
mpg['total'] = (mpg['cty']+mpg['hwy'])/2
```

```python
# mpg['total']에서 20 이상이면 pass 아니면 fail인 칼럼이름 test 열을 만들자
mpg['test'] = np.where(mpg['total']>=20, 'pass', 'fail')
mpg
```

```python
# 빈도 수 파악하기
# test 칼럼에서 pass, fail의 빈도수는 몇번일까요
a = mpg['test'].value_counts()
a
# pass    128
# fail    106
# Name: test, dtype: int64
```

```python
# 파악한 빈도수로 그림그리기
# barh = 가로로 출력 / bar = 세로출력
# a.plot(kind='barh') 

# x축 이름이 세로로 서있는게 보기 안좋으니 가로로 눕히기
a.plot(kind='bar', rot = 0 ) 
# a.plot(kind='bar')
```

```python
# total 칼럼에서 max값과 min 값을 구하세요
print(mpg['total'].max())
print(np.max(mpg['total']))
print(max(mpg['total']))

print(mpg['total'].min())
print(np.min(mpg['total']))
print(min(mpg['total']))
```

```python
# 요약을 보여주는 describe 를 사용하여 확인할 수도 있다.
mpg['total'].describe()
```

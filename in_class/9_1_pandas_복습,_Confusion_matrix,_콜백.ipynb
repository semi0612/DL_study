## Confusion matrix
![image](https://user-images.githubusercontent.com/51469989/213107635-b4e57bf5-c673-4f12-b1c1-86ccfe8ffa2b.png)

Precision(정확도; P로 검출한 것 중 실제 P의 비율) = TP / (TP+FP) <BR>
TPR(True Positive Rate) = Recall(재현율; 실제 P를 P로 예측한 비율; 예측과 실제 모두 P) =TP / (TP+FN) <BR>
Accuracy(탐지율; 맞게 검출한 비율) = (TP+TN) / (TP+FN+FP+TN) <BR>
F1 Score = 2 * `Precision` * Recall / Precision + Recall
<BR><BR>
TNR(True Negative Rate; 예측과 실제 모두 N) = TN / (TN+FP) <BR>
FPR(False Positive Rate; 실제 N인데 P로 검출) = FP / (FP+TN) <BR>
FNR(False Negative Rate; 실제 P인데 N으로 검출) = FN / (TP+FN)


## 지표(metric)
모델의 성능을 측정하는 열쇠. 특히 훈련 데이터 성능과 테스트 데이터 성능의 차이를 측정하는 것이 중요하다. 분류와 회귀에 일반적으로 사용되는 지표는 `keras.metrics` 모듈에 이미 포함되어있고, 대부분의 경우에는 여기에 포함된 지표를 사용할 것이다.<br>
케라스 지표는 `keras.metrics.Metric` 클래스를 상속한 클래스이다. 층과 마찬가지로 지표는 텐서플로우 변수에 내부 상태를 저장한다.층과 다른점이라면 이 변수가 역전파로 업데이트 되지않는다는 것으로 만약, 일반적이지 않은 작업으로 사용자 정의 지표를 만들어야 한다면 상태 업데이트 로직을 `update_state()` 메서드 안에 직접 작성해야한다.

## 콜백(callbacks)
대규모 데이터셋에서 `modle.fit()`메서드를 사용하여 수십 번의 에포크를 실행하는 것은 종이비행기를 날리는 것과 엇비슷하다. 일단 손을 떠나면 종이 비행기의 경로와 착륙 지점을 제어할 방법이 없다는 점에서 특히 그러한데, 나쁜결과로 종이비행기를 낭비하지 않으려면 대신 드론을 사용해볼 수 있다. 드론은 주변 환경을 감지하고, 감지한 데이털르 조작부테 전달하며 현재 상태를 바탕으로 자동으로 운전할 수 있다.<br>
<br>
케라스의 콜백은 fit() 호출시 모델에 전달되는 객체(특정 메서드)를 구현한 클래스 객체로, fit 메서드가 스스로 판단하고 동적으로 결정하도록 도와준다. 아래는 콜백으로 사용하는 몇 가지 예이다
- 모델 체크 포인트 저장(model checkpoint save) : 훈련하는 동안 어떤 지점에서 모델의 현재 가중치를 저장할 것인가
- 조기 종료(early stopping) : validation loss가 더 이상 향상되지 않을 때 훈련을 중지한다(= 훈련하는 동안 얻은 가장 좋은 모델의 상태에 중지)
- 훈련하는 동안 하이퍼파라미터 값을 동적으로 조정 : 옵티마이저의 학습률 같은 경우
- 훈련과 검증지표를 로그에 기록하거나 모델이 학습한 표현이 업데이트될때마다 시각화 : 우리가 fit()메서드를 사용하면서 자연스레 봤던 진행표시줄이 사실은 콜백 중 하나이다.

fit()안에 넣어사용할 수 있다. <br><br>**11월 23일에 작성했던 Early stopping 부분** 참고
```python
# early_stopping 콜백 사용
model.fit(X_train, Y_train, epoch = 1000, callbacks = [early_stopping])

# 만약 여러개의 콜백을 사용하고 싶다면 리스트 안에 넣어주면 된다.

# early_stopping과 체크포인트 사용
model.fit(X_train, Y_train, epoch = 1000, callbacks = [early_stopping, checkpoint])
```

### Model Checkpoint save
- Keras 사용 모델 트레이닝 도중에 Weights(상태) 저장하기
Keras의 콜백함수인 `ModelCheckpoint`는 모델이 학습하면서 정의한 조건을 만족했을 때 Model의 weight 값을 중간 저장해준다. 학습시간이 오래걸리는 경우 모델이 개선된 validation score를 도출해낼 때마다 weight를 중간저장해줌으로 써 혹시 중간에 끊기더라도(memory overflow나 crash가 나더라도) 다시 저장된 상태를 불러와 학습을 이어 나갈 수 있기 때문에 시간을 save해줄 수 있다.

```python
import kears

modelcheckpoint = [
    keras.callbacks.ModelCheckpoint(
      filepath, monitor='val_loss', verbose=0, save_best_only=False,
      save_weights_only=False, mode='auto', save_freq='epoch', options=None, **kwargs
    )
]
```
- filepath : 추가설명으로 만약 monitor가 val_loss일 때, 모델 경로를 '{epoch:02d} - {val_loss:.5f}.h5' 라고 입력하면 '에폭 - 해당v에폭에서의 val_loss값.h5' 로 모델이 저장된다
- monitor : 모델을 저장할 때, 기준이 되는 값. 예를 들어 validation set의 loss가 가장 작을 때 저장하고 싶으면 'val_loss'를 입력
- verbose : 0, 1 -> 1일 경우 모델이 저장될때마다 '저장되었습니다'
- save_best_only : True, False -> True인 경우 모니터되고 있는 값을 기준으로 가장 좋은 값으로 모델이 저장된다. False인 경우 매 에폭마다 모델이 filepath{epoch}으로 저장된다
- save_weights_only : True, False -> True인 경우, 모델의 weights만 저장, False인 경우, 모델 레이어 및 weights 모두 저장
- mode : 'auto', 'min', 'max' -> val_acc 인 경우, 정확도이기 때문에 클수록 좋다. 따라서 이때는 max를 입력해줘야하지만, val_loss 인 경우, loss 값이기 때문에 값이 작을수록 좋기때문에 min으로 설정해줘야한다. auto로 할 경우 알아서 min, max를 판단 후 모델을 저장
- save_freq : 'epoch' 또는 integer(정수형 숫자) -> 'epoch'을 사용할 경우, 매 에폭마다 모델 저장, integer를 사용할 경우 숫자만큼의 배치가 진행 될때 모델이 저장된다. (예를 들어 숫자 8을 입력하면, 8번째 배치가 train 된 이후, 16번째 배치가 train 된 이후 저장)
- options : tf.train.CheckpointOptions를 옵션으로 줄 수 있다. 분산환경에서 다른 디렉토리에 모델을 저장하고 싶을 경우 사용
```

### ReduceLROnPlateau : 모델 성능 향상이 없을 때 Learning Rate를 낮추는 함수
```python
reduce_lr = ReduceLROnPlateau(monitor = 'val_loss', # validation loss가 이전 epoch때보다 작아지지 않으면
                              facotr=0.2,           # 0.2배 LR을 낮춰달라
                              patience = 5,         # 5번까지는 지켜보다가 낮춰달라
                              min_lr=0.001         # 단, 0.001보다는 낮아지지 않도록 해주세요
)

model.fit(X_train, Y_train, callbacks=[reduce_lr])
```

### 텐서보드를 사용한 모니터링과 시각화 콜백

- 훈련하는 동안 측정 지표를 시각적으로 모니터링
- 모델 구조를 시각화
- 활성화 출력과 그레이디언트의 히스토그램 그리기
- 임베딩을 3D로 표현

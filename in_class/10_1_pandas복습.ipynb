### pandas 복습

```python
# 데이터 프레임 만들기

# 국어	영어	수학
# A	80	90	100
# B	70	80	90

import pandas as pd

pd.DataFrame([[80,90,100],[70,80,90]],
             columns = {'국어', '영어', '수학'},
             index = {'A', 'B'})
```

```python
pd.DataFrame(data={'국어': [80, 70], '영어': [90, 80], '수학':[100, 90]}, index = ['A', 'B'])
```

```python
# 위 dataframe 에 과학 칼럼 (90, 90)을 추가해보세요
df = pd.DataFrame(data={'국어': [80, 70], '영어': [90, 80], '수학':[100, 90]}, index = ['A', 'B'])
df['과학'] = [90, 90]
df
```

```python
# 수학 칼럼을 삭제해보세요
df.drop(columns='수학', inplace=True)
df
```

```python
# 총점 칼럼을 추가
df['총점'] =  df['국어']+df['영어']+df['과학']
df
```

```python
# 행 이름이 c이고 70, 80, 90, 240인 행을 추가해보세요
df.loc['c'] = [70, 80, 90, 240]
df
```

```python
# 인덱스 명 조작
# c를 C로 변경
df.index.values[len(df)-1] = 'C'
df
```

```python
# 총점의 260을 결측치로 만들어주세요
import numpy as np

df['총점'].replace(260, np.nan, inplace=True)
df
```

```python
df['총점'].replace(np.nan, 260, inplace=True)
df
```

```python
# 평균이 85점 이상이면 A+ 그 이하면 A-로 표현하는 열을 만들어주세요
cond1 = df['총점']/3 >= 85
df['학점'] = np.where(cond1, 'A+', 'A-' )
df
```

```python
# 총점, 국어 점수의 오름차순으로 정렬해주세요
df.sort_values(by=['총점', '국어'])
```

```python
# 연도별 cty의 평균은
mpg.groupby(['year'])[['cty']].mean()
```

```python
# 아우디의 연도별 cty의 평균은
mpg[mpg['manufacturer'] == 'audi'].groupby('year')['cty'].agg(np.mean)
```
